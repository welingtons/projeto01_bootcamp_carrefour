```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Qual a receita total?
df["Valor Venda"].sum()
```

```python
# Qual o custo total?
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])  # Criando a coluna de custo
```

```python
df.head(1)
```

```python
# Qual o custo total
round(df['custo'].sum(), 2)
```

```python
# Agora que tenos a receita e o custo e o total, podemos achar o lucro total
# Vamos criar uma coluna de lucro que será Receita - Custo
df['Lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(20)
```

```python
# Total lucro
round(df['Lucro'].sum(), 2)
```

```python
# Criando uma coluna com total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
# Visualizando as 10 primeiras linhas
df.head(10)
```

### Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em numérica

```python
# Extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(10)
```

```python
# Verificando o tipo da coluna Tempo_envio
df['Tempo_envio'].dtype
```

```python
# Média do tempo de envio por Marca
round(df.groupby('Marca')['Tempo_envio'].mean(), 2)
```

### Missing Values

```python
# Verificando se temos dados faltantes
df.isnull().sum()
```

### E, se a gente quiser o Lucro por Ano e Por Marca

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Vamos Agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gráfico Total de Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head(10)
```

```python
df_2009.sample(10)
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca', color='red')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro x Classe', color='orange')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
